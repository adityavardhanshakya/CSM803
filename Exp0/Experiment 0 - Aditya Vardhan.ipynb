```python
import numpy as np

# Generate a random 5x5 matrix with entries between 0 and 1
random_matrix = np.random.rand(5, 5)

# Find the largest and smallest entries of the matrix
largest_entry = np.max(random_matrix)
smallest_entry = np.min(random_matrix)

print("Largest entry:", largest_entry)
print("Smallest entry:", smallest_entry)
```

```python
import numpy as np

# Generate a vector of 100,000 random numbers
x = np.random.rand(100000)

# Find the maximum and minimum values
max_value = np.max(x)
min_value = np.min(x)

print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a vector of 100,000 random numbers
x = np.random.rand(100000)

# Plot the histogram with 100 bins
plt.hist(x, bins=100)
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a vector of 100,000 random numbers from a normal distribution
x_normal = np.random.randn(100000)

# Plot the histogram with 100 bins
plt.hist(x_normal, bins=100)
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random vectors with different numbers of elements
x1 = np.random.randn(5)
x2 = np.random.randn(50)
x3 = np.random.randn(500)
x4 = np.random.randn(5000)
x5 = np.random.randn(50000)
x6 = np.random.randn(500000)

# Plot histograms for each vector with 100 bins
plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.hist(x1, bins=100)
plt.title('Histogram of x1 (5 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(3, 2, 2)
plt.hist(x2, bins=100)
plt.title('Histogram of x2 (50 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(3, 2, 3)
plt.hist(x3, bins=100)
plt.title('Histogram of x3 (500 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(3, 2, 4)
plt.hist(x4, bins=100)
plt.title('Histogram of x4 (5000 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(3, 2, 5)
plt.hist(x5, bins=100)
plt.title('Histogram of x5 (50000 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(3, 2, 6)
plt.hist(x6, bins=100)
plt.title('Histogram of x6 (500000 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
