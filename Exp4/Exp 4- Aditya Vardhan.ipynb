```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define patterns (AND mapping)
X = np.array([[1, 0, 0],  # Patterns to be learned. One pattern per row.
              [1, 0, 1],
              [1, 1, 0],
              [1, 1, 1]])
d = np.array([0, 0, 0, 1])  # Desired outputs for AND classification

# Initialize variables
eta = 1.0  # Learning rate
W = np.array([0, 0, 0])  # Initialize weight vector
updated = 1  # Update flag checks for exit
```

```python
# Perceptron Learning Algorithm
W_matrix = np.zeros((1, 3))
while updated == 1:
    for k in range(4):  # For each training pattern
        y = np.dot(X[k], W)  # Compute activation
        if y >= 0 and d[k] == 0:
            W = W - eta * X[k]
            update = 1  # We made a change on pattern k
        elif y <= 0 and d[k] == 1:
            W = W + eta * X[k]
            update = 1  # We made a change on pattern k
        else:
            update = 0  # We made no change on pattern k
        W_matrix = np.vstack([W_matrix, W])  # Record the weights at each iteration
    
    flag = np.dot(update, update)  # Check if any pattern required update
    if flag > 0:
        updated = 1
    else:
        updated = 0
```

```python
# Plot weight trajectory in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(W_matrix[:, 0], W_matrix[:, 1], W_matrix[:, 2], 'b-^')
ax.grid(True)
ax.view_init(-68, 58)
ax.set_title('Perceptron Learning Algorithm Weight Space View')
ax.set_xlabel('Weight 1')
ax.set_ylabel('Weight 2')
ax.set_zlabel('Weight 3')
plt.show()
```
