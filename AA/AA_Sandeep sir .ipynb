```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate patterns
def generate_patterns():
    x = np.linspace(-2, 2, 25)
    y = np.linspace(-2, 2, 25)
    pattern = []
    for i in range(25):
        for j in range(25):
            pattern.append([x[i], y[j], np.exp(-(x[i]**2 + y[j]**2))])
    pattern = np.array(pattern)
    np.random.shuffle(pattern)
    return pattern

# Function to split data into train and test sets
def split_data(pattern):
    trainpattern = pattern[:500, :]  # 80% train
    testpattern = pattern[500:, :]  # 20% test
    return trainpattern, testpattern

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Neural network training function
def train_neural_network(trainpattern, testpattern, architecture, eta, alpha):
    Q = trainpattern.shape[0]  # Number of patterns
    n = trainpattern.shape[1] - 1  # Input layer size
    p = 1  # Output layer size
    q = architecture[1]  # Hidden layer size

    # Initialize weights
    Wih = 2 * np.random.rand(n + 1, q) - 1
    Whj = 2 * np.random.rand(q + 1, p) - 1

    # Add bias to input and hidden layers
    trainpattern_with_bias = np.hstack((np.ones((Q, 1)), trainpattern[:, :2]))
    testpattern_with_bias = np.hstack((np.ones((testpattern.shape[0], 1)), testpattern[:, :2]))

    # Initialize momentum terms
    DeltaWihOld = np.zeros((n + 1, q))
    DeltaWhjOld = np.zeros((q + 1, p))

    # Training parameters
    epoch = 0
    tol = 0.01
    test_error = 2 * tol
    train_errors = []
    test_errors = []

    while test_error > tol:
        epoch += 1
        train_error = 0
        for k in range(Q):
            # Forward pass
            Zh = np.dot(trainpattern_with_bias[k, :], Wih)
            Sh = np.hstack((1, sigmoid(Zh)))
            Yj = np.dot(Sh, Whj)
            Sy = sigmoid(Yj)

            # Backpropagation
            Ek = trainpattern[k, 2] - Sy
            deltaO = Ek * sigmoid_derivative(Sy)
            DeltaWhj = np.outer(Sh, deltaO)
            deltaH = np.dot(Whj[1:, :], deltaO) * sigmoid_derivative(Sh[1:])
            DeltaWih = np.outer(trainpattern_with_bias[k, :], deltaH)

            # Update weights
            Wih += eta * DeltaWih + alpha * DeltaWihOld
            Whj += eta * DeltaWhj + alpha * DeltaWhjOld
            DeltaWihOld = DeltaWih
            DeltaWhjOld = DeltaWhj
            train_error += Ek ** 2

        train_error /= Q
        train_errors.append(train_error)

        # Calculate test error
        test_error = 0
        for l in range(testpattern.shape[0]):
            Zh = np.dot(testpattern_with_bias[l, :], Wih)
            Sh = np.hstack((1, sigmoid(Zh)))
            Yj = np.dot(Sh, Whj)
            Sy = sigmoid(Yj)
            test_error += (testpattern[l, 2] - Sy) ** 2

        test_error /= testpattern.shape[0]
        test_errors.append(test_error)

        print(f'Epoch {epoch}: Training Error = {train_error}, Testing Error = {test_error}')

    return epoch, train_errors, test_errors

# Main function
def main():
    # Initialize parameters
    trials = 8
    architectures = [[2, 3, 1], [2, 4, 1], [2, 5, 1], [2, 6, 1], [2, 3, 2], [2, 4, 2], [2, 5, 2], [2, 6, 2]]
    etas = [0.8, 0.5, 1, 1, 1, 1, 1, 1]
    alphas = [0.01, 0.01, 0.01, 0.1, 0.02, 0.03, 0.1, 0.7]

    # Generate patterns
    pattern = generate_patterns()

    # Split data
    trainpattern, testpattern = split_data(pattern)

    # Train the network and record observations
    observations = []
    for trial in range(trials):
        architecture = architectures[trial]
        eta = etas[trial]
        alpha = alphas[trial]
        print(f"Trial {trial + 1}: Architecture={architecture}, Learning Rate={eta}, Momentum={alpha}")
        epoch, train_errors, test_errors = train_neural_network(trainpattern, testpattern, architecture, eta, alpha)
        observations.append([trial + 1, architecture, eta, alpha, epoch, train_errors[-1], test_errors[-1]])

        # Plot error curves
        plt.plot(train_errors, label='Train Error')
        plt.plot(test_errors, label='Test Error')
        plt.title(f'Trial {trial + 1} Errors')
        plt.xlabel('Epochs')
        plt.ylabel('Error')
        plt.legend()
        plt.show()

    # Display observations
    print("\nObservations:")
    print("S.No\tArchitecture\tLearning Rate\tMomentum\tFinal Epochs\tFinal Training Error\tFinal Testing Error")
    for obs in observations:
        print("\t".join(str(item) for item in obs))

if __name__ == "__main__":
    main()
```
