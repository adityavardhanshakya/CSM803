```python
import numpy as np
import matplotlib.pyplot as plt

# Assume 200 data points
max_points = 200

# Generate the x linspace
x = np.linspace(0, 2.5, max_points)

# Define a straight line
y = 0.5 * x + 0.333

# Generate scatter vector
scatter = np.random.rand(max_points)

# Compress scatter to 0.1
ep = 0.1

# Set up desired values
d = ((2 * scatter - 1) * ep) + y

# Set learning rate
eta = 0.01

# Randomize weights
w = 3 * (2 * np.random.rand(2) - 1)

# Train for 50 epochs
for loop in range(50):
    # Randomize order
    randindex = np.random.permutation(max_points)
    
    # For each data point
    for j in range(max_points):
        # Get the index
        i = randindex[j]
        
        # Compute signal value
        s_i = w[0] + w[1] * x[i]
        
        # Compute pattern error
        err_i = d[i] - s_i
        
        # Change the weights
        w[0] += eta * err_i / (1 + x[i]**2)
        w[1] += eta * err_i * x[i] / (1 + x[i]**2)

# Finally compute the function learnt
s = w[0] + w[1] * x

# Plot the result
plt.plot(x, s, 'r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
max_points = 200
eta = 0.01
epochs = [5, 8, 10]

# Generate the x linspace
x = np.linspace(0, 2.5, max_points)

# Define a straight line
y = 0.5 * x + 0.333

# Generate scatter vector
scatter = np.random.rand(max_points)

# Compress scatter to 0.1
ep = 0.1

# Set up desired values
d = ((2 * scatter - 1) * ep) + y

# Plot the scatter data
plt.plot(x, d, 'b.')

# Set learning rate
w = 3 * (2 * np.random.rand(2) - 1)  # Randomize weights

# Train for each specified number of epochs
for epoch in epochs:
    for _ in range(epoch):
        # Randomize order
        randindex = np.random.permutation(max_points)

        # For each data point
        for j in range(max_points):
            # Get the index
            i = randindex[j]

            # Compute signal value
            s_i = w[0] + w[1] * x[i]

            # Compute pattern error
            err_i = d[i] - s_i

            # Change the weights
            w[0] += eta * err_i / (1 + x[i]**2)
            w[1] += eta * err_i * x[i] / (1 + x[i]**2)

    # Finally compute the function learnt
    s = w[0] + w[1] * x

    # Plot the result
    plt.plot(x, s, label=f'Epochs: {epoch}')

# Show plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Effect of Reducing Epochs on LMS Learning')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
max_points = 200
eta = 0.01
epoch = 50

# Generate the x linspace
x = np.linspace(0, 2.5, max_points)

# Define a straight line
y = 0.5 * x + 0.333

# Generate scatter vector
scatter = np.random.rand(max_points)

# Compress scatter to 0.1
ep = 0.1

# Set up desired values
d = ((2 * scatter - 1) * ep) + y

# Plot the scatter data
plt.plot(x, d, 'b.', label='Data Scatter')

# Set learning rate
w = 3 * (2 * np.random.rand(2) - 1)  # Randomize weights

# Train for 50 epochs
for _ in range(epoch):
    # Randomize order
    randindex = np.random.permutation(max_points)

    # For each data point
    for j in range(max_points):
        # Get the index
        i = randindex[j]

        # Compute signal value
        s_i = w[0] + w[1] * x[i]

        # Compute pattern error
        err_i = d[i] - s_i

        # Change the weights
        w[0] += eta * err_i / (1 + x[i]**2)
        w[1] += eta * err_i * x[i] / (1 + x[i]**2)

    # Finally compute the function learnt
    s = w[0] + w[1] * x

    # Plot the hyperplane during learning process
    plt.plot(x, s, 'r-', alpha=0.1)

# Plot the final hyperplane
plt.plot(x, s, 'g-', linewidth=1, label='Final Hyperplane')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperplane during learning process')
plt.show()
```
