```python
from sklearn.datasets import load_iris

# Load Fisher Iris dataset
iris = load_iris()
x = iris.data  # Features (sepal length, sepal width, petal length, petal width)
labels = iris.target  # Labels (setosa: 0, versicolor: 1, virginica: 2)
```

```python
import matplotlib.pyplot as plt

# Plot the data in the x3-x4 subspace
plt.figure()
plt.plot(x[:50, 2], x[:50, 3], 'b.', label='Setosa')
plt.plot(x[50:100, 2], x[50:100, 3], 'g.', label='Versicolor')
plt.plot(x[100:150, 2], x[100:150, 3], 'r.', label='Virginica')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Data in x3-x4 Subspace')
plt.legend()
plt.show()
```

```python
from scipy.spatial import ConvexHull

# Compute convex hulls
k_setosa = ConvexHull(x[:50, 2:4])
k_versicolor = ConvexHull(x[50:100, 2:4])
k_virginica = ConvexHull(x[100:150, 2:4])

# Plot the data with convex hulls in the x3-x4 subspace
plt.figure()
plt.plot(x[:50, 2], x[:50, 3], 'b.', label='Setosa')
plt.plot(x[50:100, 2], x[50:100, 3], 'g.', label='Versicolor')
plt.plot(x[100:150, 2], x[100:150, 3], 'r.', label='Virginica')

# Plot convex hulls
for simplex in k_setosa.simplices:
    plt.plot(x[:50, 2][simplex], x[:50, 3][simplex], 'b-', linewidth=1)
for simplex in k_versicolor.simplices:
    plt.plot(x[50:100, 2][simplex], x[50:100, 3][simplex], 'g-', linewidth=1)
for simplex in k_virginica.simplices:
    plt.plot(x[100:150, 2][simplex], x[100:150, 3][simplex], 'r-', linewidth=1)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Data with Convex Hulls in x3-x4 Subspace')
plt.legend()
plt.show()
```
